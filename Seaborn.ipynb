## Visualization with Seaborn - Appearance of plots - Figure Aesthetic

### Continous Data Plotting
   - Infinite number of possible values within a selected range

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)                                                  # Plot below is from 0 to 14, 100 is no of samples. Change 100 to 10 to understand
   for i in range(1, 5):                                                        # Makes 4 plots
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)                          # Making plots

sinplot()                                                                       # Calls the sinplot function
plt.show()                                                                      # Plotting using pyplot in matplotlib
```

```python
sns.set()                                                                       # Set Seaborn defaults
sinplot()                                                                       # call the sinplot function defined in above cell
plt.show()                                                                      # Show the plot
```

## Categorical Data plotting
- Variable under study is categorical. Can take on one of a limited, and usually fixed number of possible values, assigning each individual or other unit of observation to a particular group or nominal category on the basis of some qualitative property.

- Plots used are
    1. stripplot()
    2. swarmplot()

### 1. stripplot()
   - used when one of the variable under study is categorical
   - represents the data in sorted order along any one of the axis

```python
df = sns.load_dataset('iris')
sns.stripplot(x = "species", y = "petal_length", data = df, jitter = False, color='r')   # Jitter = False -> Points overlap
                                                                                # so difficult to see all points
plt.show()
```

```python
df = sns.load_dataset('iris')
sns.stripplot(x = "species", y = "petal_length", data = df, jitter = True, color='r')   # Jitter = True -> adds noise for better visualization of distribution of points
plt.show()
```

### 2. swarmplot()
   - Alternative to jitter function in Stirpplot
   - positions each point of scatter plot on the categorical axis and thereby avoids overlapping points

```python
df = sns.load_dataset('iris')
sns.swarmplot(x = "species", y = "petal_length", data = df, color = "Red")
plt.show()
```

## Visualizing linear relationships
   - Many datasets contain multiple quantitative variables, and the goal of an analysis is often to relate those variables to each other.
   - xploring a dataset through visualization quick and easy
   - There are two main functions in Seaborn to visualize a linear relationship determined through regression.
   - Makes a scatter plot and fits a linear regression line
       - regplot() : accepts the x and y variables in a variety of formats including simple numpy arrays, pandas Series objects, or as references to variables in a pandas DataFrame
       - lmplot() : accepts data as a required parameter and the x and y variables must be specified as strings. This data format is called “long-form” data. Implot is combination of regplot and facetgrid(we see below)

```python
df = sns.load_dataset('tips')
print (df.head())

sns.regplot(x = "total_bill", y = "tip", data = df,  scatter_kws={"color": "black"}, line_kws={"color": "red"})
sns.lmplot(x = "total_bill", y = "tip", data = df,  scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.show()
```

```python
df = sns.load_dataset('tips')
print (df.head())

sns.lmplot(x = "size", y = "tip", data = df,  scatter_kws={"color": "black"}, line_kws={"color": "red"})  # size has a discrete values
plt.show()
```

```python
df = sns.load_dataset('anscombe')                                               # Dataset published by statistician Francis Anscombe in 1973 having 11 x y points to show
                                                                                # importance of plotting dataset before analyzing, and show effect of outliers
print (df.head(5))

sns.lmplot(x="x", y="y", data=df.query("dataset == 'I'"),  scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.show()

# data is linear and can be shown
```

```python
df = sns.load_dataset('anscombe')
print (df.query("dataset == 'II'").head())

sns.lmplot(x = "x", y = "y", data = df.query("dataset == 'II'"),  scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.show()

# data is non-linear (data takes high deviation) and the above methods cannot generalize the regression line.
```

```python
df = sns.load_dataset('anscombe')
print (df.query("dataset == 'II'").head())

sns.lmplot(x = "x", y = "y", data = df.query("dataset == 'II'"),order = 2,  scatter_kws={"color": "black"}, line_kws={"color": "red"})
                                                                                # Since data is not linear, use order > 1
plt.show()
```

### FacetGrid :
   - Helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels.

   - Up to three dimensions − row, col, and hue
      - row, col - x and y axis
      - hue - Depth of axis - different colours

   - FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid.

```python
df = sns.load_dataset('tips')
print (df.head())

g = sns.FacetGrid(df.query("smoker == 'No' or smoker == 'Yes'"), col = "time", hue = "smoker", hue_kws = {'color': ['r', 'b']})
                                                                                # 2 plots as col = "time" has 2 values Lunch, Dinner
g.map(plt.hist, "tip")                                                          # x axis = tip
plt.show()

```

```python
df = sns.load_dataset('tips')
print (df.head())

g = sns.FacetGrid(df.query("smoker == 'No' or smoker == 'Yes'"), col = "sex", hue = "smoker", hue_kws={'color': ['r', 'b']})
                                                                                # 2 plots as col = "sex" has 2 values Male, Female
                                                                                # hue to distinguish smoker/non smoker
g.map(plt.scatter, "total_bill", "tip")                                         # x axis -> total bill, y axis -> tip
                                                                                # Scatter graph
plt.show()
```

### PairGrid /scatterplot matrix:
   - Uses different pair of variable for each subplot.
   - Forms a matrix of sub-plots.
   

```python
df = sns.load_dataset('tips')
print (df.head())

g = sns.PairGrid(df, hue="sex")
g.map(plt.scatter);
plt.show()
```

```python
df = sns.load_dataset('iris')
print(df.head())


g = sns.PairGrid(df)
g.map_diag(plt.hist)                                                            # Diagonal plot are histogram
g.map_offdiag(plt.scatter)                                                      # other than diagonal is scatter plot
plt.show()
```

```python
df = sns.load_dataset('iris')
print (df.head())

g = sns.PairGrid(df)
g.map_upper(plt.scatter)                                                        # upp of diagonal is scatter plot
g.map_lower(sns.kdeplot, cmap = "Blues_d")                                      # lower of diagonal is  Kernel Density Estimate plot

g.map_diag(plt.hist)                                                            # Diagonal is Histogram

plt.show()
```
